```python
def convert_to_yolo_format(input_file, output_directory, image_width, image_height):
    with open(input_file, 'r') as f:
        lines = f.readlines()

    annotations = {}

    # Collect annotations based on video_id and frame
    for line in lines:
        parts = line.strip().split(',')
        video_id, frame, tl_x, tl_y, bb_width, bb_height, class_id = parts

        # Convert video_id and frame to padded strings
        video_id_str = str(video_id).zfill(3)
        frame_str = str(frame)  # No zfill needed for frame ID

        # Create filename based on video_id and frame
        filename = f"{video_id_str}_{frame_str}.txt"

        # Calculate center coordinates and normalize them
        center_x = (float(tl_x) + float(bb_width) / 2) / image_width
        center_y = (float(tl_y) + float(bb_height) / 2) / image_height
        width = float(bb_width) / image_width
        height = float(bb_height) / image_height

        # Merge class IDs into fewer classes
        if class_id in ['2', '4', '6', '8']:
            class_index = 1  # RiderHelmet
        elif class_id in ['3', '5', '7', '9']:
            class_index = 2  # RiderNoHelmet
        elif class_id == '1':
            class_index = 0  # Motorbike
        else:
            continue  # Skip other classes

        # Prepare annotation line
        annotation_line = f"{class_index} {center_x} {center_y} {width} {height}\n"

        # Append annotation to corresponding file
        if filename not in annotations:
            annotations[filename] = []
        annotations[filename].append(annotation_line)

    # Write annotations to files
    for filename, annotation_lines in annotations.items():
        with open(output_directory + "/" + filename, 'w') as f:
            f.writelines(annotation_lines)

# Example usage
input_file = '/home/ss6928/E6691_assign2/labels/gt_training.txt'
output_directory = '/home/ss6928/E6691_assign2/labels/labels_training_yolo'
image_width = 1920  # Update with actual image width
image_height = 1080  # Update with actual image height

convert_to_yolo_format(input_file, output_directory, image_width, image_height)

```

```python
import cv2
import os

# Function to extract frames from a video
def extract_frames(video_path, output_directory):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Create output directory if it doesn't exist
    os.makedirs(output_directory, exist_ok=True)

    # Initialize frame count
    frame_count = 0

    # Read until video is completed
    while cap.isOpened():
        # Read a frame from the video
        ret, frame = cap.read()

        if not ret:
            break

        # Save the frame as a JPG file
        frame_count += 1
        frame_filename = f"{output_directory}/{os.path.splitext(os.path.basename(video_path))[0]}_{frame_count}.jpg"
        cv2.imwrite(frame_filename, frame)

    # Release the video capture object
    cap.release()

# Set paths
video_directory = '/home/ss6928/E6691_assign2/Videos_trainin_val/training_videos'
output_directory = '/home/ss6928/E6691_assign2/images/train'

# Loop through each video and extract frames
for video_name in os.listdir(video_directory):
    video_path = os.path.join(video_directory, video_name)

    extract_frames(video_path, output_directory)

print("Frame extraction completed.")
```

```python
import os

# Path to the dataset folder
dataset_folder = "/home/ss6928/E6691_assign2/Data/train"

# Function to remove files
def remove_files(files, folder):
    for file_name in files:
        file_path = os.path.join(folder, file_name)
        if os.path.exists(file_path):
            os.remove(file_path)
            print(f"Removed: {file_path}")
        else:
            print(f"File not found: {file_path}")

# Function to get image and label file paths
def get_file_paths(dataset_folder):
    image_folder = os.path.join(dataset_folder, "images")
    label_folder = os.path.join(dataset_folder, "labels")

    image_files = [f for f in os.listdir(image_folder) if f.startswith("036")]
    label_files = [f for f in os.listdir(label_folder) if f.startswith("036")]

    return image_files, label_files

# Get image and label file paths
image_files, label_files = get_file_paths(dataset_folder)

# Remove problematic images
remove_files(image_files, os.path.join(dataset_folder, "images"))

# Remove corresponding labels
remove_files(label_files, os.path.join(dataset_folder, "labels"))
```
